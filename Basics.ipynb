Calculating the statistical values for month , year and postal code have no sense so we need to exclude these columns from our statistical analysis. Also there's a column that all values are zeros called nr_pictures of course wedon't need to calculate anything for this column

There are columns we need to include in our statistical calculations are not included here which means that they are stored as text so we need to turn them into numeric colunms

```python
autos['price']= autos['price'].str.replace('$','')
autos['price']= autos['price'].str.replace(',','')
autos['odometer']= autos['odometer'].str.replace('km','')
autos['odometer']= autos['odometer'].str.replace(',','')
```

```python
autos.rename({'odometer':'odometer_km'})
autos['price']=pd.to_numeric(autos['price'])
autos['odometer']=pd.to_numeric(autos['odometer'])
```

```python
brands = autos['brand'].value_counts().reset_index()
```

```python
brands
```

```python
prec=[]
for i in range(len(brands)):
    prec.append((brands.iloc[i]['brand']/len(autos['brand']))*100)
brands['precentage']=prec
```

```python
brands
```

```python
top_brands = brands.loc[brands['precentage']>5]
top_brands
```

```python
autos.column
```

```python
groups = autos.groupby('brand')
brands_avg_price={}
for name,group in groups:
    if name in top_brands['index'].values:
        brands_avg_price[name]=(group['price'].mean(),group['mileage'].mean())
```

```python
print(brands_avg_price)
```

From the top brands we have here we can see that the average price between 2974 to 3593 dollars and that the cheapest brand of them all is OPEL and the most expensive brand of them all is BMW
